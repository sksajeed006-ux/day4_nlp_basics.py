```python
#Sentence Tokenization
from nltk.tokenize import sent_tokenize #importing method from nltk
import nltk
#downloading of model
nltk.download('punkt_tab')
text = "Artificial Intelligence is transforming industries. NLP is a key part of AI. Tokenization is the first NLP step."
print(sent_tokenize(text))

```

```python
#word tokenization
from nltk.tokenize import word_tokenize #importing method from nltk
import nltk
text = "Artificial Intelligence is transforming industries. NLP is a key part of AI. Tokenization is the first NLP step."
print(word_tokenize(text))

```

```python
#Stemming with Porter Stemmer
from nltk.stem import PorterStemmer #importing method from package

stemmer = PorterStemmer() #defining object
words = ["playing", "played", "plays", "happily", "studies"]
for word in words:
   stemmed_words = stemmer.stem(word)
   print(stemmed_words)
```

```python
#Lemmatization with WordNet
from nltk.stem import WordNetLemmatizer #importing method from package
import nltk
nltk.download('wordnet') #downloaded model
lemmatizer = WordNetLemmatizer() #defining object
words = ["running", "cars", "better", "children", "feet"]
for word in words:
  lemmatized_word  = lemmatizer.lemmatize(word)
  print(lemmatized_word)
```

```python
#Stemming with Porter Stemmer
from nltk.stem import PorterStemmer , WordNetLemmatizer #importing method from package
import nltk
nltk.download('wordnet') #downloaded model

stemmer = PorterStemmer() #defining object
lemmatizer = WordNetLemmatizer() #defining object

words = ["running", "cars", "better", "children", "feet"]
for word in words:
   stemmed_words = stemmer.stem(word)
   lemmatized_word  = lemmatizer.lemmatize(word)

   print("word:",word, "|stem:", stemmed_words ,"|lemma:",lemmatized_word )
```

```python
#Lemmatization with WordNet
from nltk.stem import WordNetLemmatizer #importing method from package
import nltk
nltk.download('wordnet') #downloaded model
lemmatizer = WordNetLemmatizer() #defining object
words = ["running", "cars", "better", "children", "feet"]
for word in words:
  lemmatized_word  = lemmatizer.lemmatize(word,"v")
  print(lemmatized_word)
```
